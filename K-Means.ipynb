El algoritmo conocido como **K-Vecinos más cercanos** es un proceso iterativo en el que a partir de los datos de entrada se entrena un modelo en el que de manera orgánica se ajustan a partir de centroides, que en principio, son definidos de forma aleatoria.

### Ejemplo guiado

```python
# Importamos las librerías básicas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn')
```

```python
# Importamos una función para crear nubes de puntos
from sklearn.datasets import make_blobs
```

```python
# Creamos los datos para nuestro ejemplo
X, y = make_blobs(n_samples=200, centers=4, cluster_std=0.55, random_state=0)
X.shape
```

```python
y.shape
```

```python
# Graficamos en un gráfico de ScatterPlot para ver la dispersión de los datos
plt.scatter(X[:, 0], X[:, 1])
plt.title('Diagrama de dispersión', fontsize=18)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Importamos el algoritmo 
from sklearn.cluster import KMeans
```

```python
# Definimos el número de clusters que queremos ajustar
kmeans_model = KMeans(n_clusters=4)
```

```python
# Corremos el modelo con los datos graficados y hacemos una predicción
kmeans_model.fit(X)

y_kmeans_model = kmeans_model.predict(X)
```

```python
# Graficamos separando los clusters por color y agregamos los centroides calculados
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans_model, s=50, cmap='viridis')
centers = kmeans_model.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c=['c', 'red', 'yellow', 'green'], s=200, alpha=0.75)
plt.title('Clusterización por K-Means', fontsize=18)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
